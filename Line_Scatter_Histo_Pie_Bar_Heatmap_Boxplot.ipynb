```python
store=pd.read_csv("store.csv")
```

```python
store.head()
```

```python
s1=store.groupby('Location')['TotalSales'].sum().sort_values(ascending=False)
```

```python
s1
```

```python
plt.pie(s1,labels=s1.index,autopct="%0.2f%%")
plt.show()
```

## Bar Chart

```python
plt.figure(figsize=(10,6))
plt.bar(s1.index,s1,color=["green","blue","yellow","red"],width=0.4)
plt.title("Sales by Location")
plt.xlabel("Location")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.barh(s1.index,s1,color=["green","blue","yellow","red"])
plt.title("Sales by Location")
plt.xlabel("Location")
plt.ylabel("Total Sales")
plt.show()
```

```python
store.head()
```

```python
s3=store.groupby(["Location","StoreType"])['TotalSales'].sum()
```

```python
s3.index
```

```python
s4=s3.unstack()
```

```python
s4
```

```python
s4.index
```

```python
s4['Apparel']
```

```python
sm=s4.Apparel+s4.Electronincs
```

```python
plt.figure(figsize=(10,6))
plt.bar(s4.index,s4['Apparel'],label="Apparel")
plt.bar(s4.index,s4['Electronincs'],bottom=s4['Apparel'],label="Electronics")
plt.bar(s4.index,s4['Super Market'],bottom=sm,label="Super Market")
plt.legend()
plt.show()
```

## Clustered bar chart or side by side bar chart

```python
plt.figure(figsize=(10,6))
sns.set_theme(style="darkgrid")
sns.barplot(x='Location',y='TotalSales',data=store,estimator=sum,hue="StoreType",palette="hot",edgecolor="black",hatch="/",errorbar=None)
plt.legend()
plt.show()
```

```python
store.groupby("Location")["TotalSales"].sum()
```

## Heat Map

```python
plt.figure(figsize=(15,8))
```

```python
store.info()
```

```python
store.select_dtypes(exclude='number').columns # display non numberic columns
```

```python
non_numeric_columns = store.select_dtypes(exclude='number').columns 
```

```python
numeric_store = store.drop(columns=non_numeric_columns)
```

```python
numeric_store.corr()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(numeric_store.corr(),cmap="RdBu",annot=True,vmin=-1,vmax=1,annot_kws={"fontsize":16,"fontweight":"bold","fontfamily":"serif"},linewidth=1,linecolor="black")
plt.show()
```
