### Question 1 : To create a scatter graph using a random distribution in X and Y, you will need to write a program in Python. Plot the data against each other. (You can use randn () )

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.figure(figsize=(5,3))

x = np.random.randn(50)
y = np.random.randn(50)

plt.title('Random Distribution in X and Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(x,y, label='XY')
plt.legend(loc='upper right')
plt.show()
```

### Question 2:  Perform Following task on given dataset. 

## Task 1: Read the data

```python
ds=pd.read_csv('Dataset_mat.csv')
ds.head()
```

### Task 2 : Display it using a line plot and read out the total profit for all of the months.

```python
plt.figure(figsize=(5,3))

x=ds.Month
y=ds.Total_profit

plt.title('Total Profit for all the Months')
plt.xlabel('Months')
plt.ylabel('Total Profit')
plt.plot(x,y, label='Total Profit')
plt.legend(loc='lower right')
plt.show()
```

### Task 3: Determine the cumulative profit for all months and display it as a line plot using the following style properties. 
It is recommended that the line style be dotted and the line color be red.
Display the legend in the bottom right hand corner of the screen.
Name of the label X = month number
Y-label name = total profit
Put a circle marker
As read, the color of the line marker The line width should be 3

```python
ds['cum_profit']= ds.Total_profit.cumsum()
ds.head()
```

```python
plt.figure(figsize=(5,3))

x=ds.Month
y=ds.cum_profit

plt.plot(x,y, '--',color='red', marker='o', label='Cumulative Profit', linewidth=3)
plt.legend(loc='lower right')
plt.title('Cumulative Profit for all Months')
plt.xlabel('Months Number')
plt.ylabel('Profit')
plt.show()
```

### Task 4: Examine the monthly sales figures of toothpaste and present them in the form of a scatter plot.


```python
plt.figure(figsize=(5,3))

x=ds.Month
y=ds.Toothpaste

plt.title('Monthly Sales of Toothpaste')
plt.xlabel('Months')
plt.ylabel('Toothpaste')
plt.scatter(x,y, label='Toothpaste')
plt.legend(loc='lower right')
plt.show()
```

### Task 5 : Read the monthly total profit and plot it on the histogram so that you can observe the most typical profit ranges.


```python
plt.figure(figsize=(10,3))

y=ds.Total_profit

plt.hist(y, label='Profit', edgecolor='black', color='yellow', bins=[14000,14500,15000,15500,16000,16500,17000,17500,18000])
plt.xticks([14000,14500,15000,15500,16000,16500,17000,17500,18000])
plt.title('Total Profit Frequency')
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```
