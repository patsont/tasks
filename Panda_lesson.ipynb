## Handling Missing Values

```python
# NaN : Missing value
weather=pd.read_csv("weather_data1 (1).csv")
```

```python
weather
```

```python
# Identying missing values
weather.isnull().sum()
```

```python
weather.notnull().sum()
```

```python
# Missing value imputation : Replacing missing values with sensible information
weather.fillna(0)
```

```python
weather.fillna({"temperature":0,"windspeed":0,"event":"No Event"})
```

```python
weather.fillna({"temperature":weather.temperature.mean(),"windspeed":weather.windspeed.mean(),"event":"No Event"})
```

```python
weather.ffill()
```

```python
weather.bfill()
```

## Concat, Merge,Append

```python
city1=pd.DataFrame({"City":["Texas","Montana","Illinois"],"Temperature":[32,45,30],"Humidity":[80,60,78]})
```

```python
city1
```

```python
city2=pd.DataFrame({"City":["New York","Chicago","California"],"Temperature":[21,14,35],"Humidity":[68,60,75]})
```

```python
city2
```

```python
city=pd.concat([city1,city2],ignore_index=True)
```

```python
city
```

```python
df5=pd.DataFrame({"City":["New York","Chicago","California"],"Humidity":[80,60,78]})
```

```python
df5
```

```python
df6=pd.DataFrame({"City":["New York","California","Illinois"],"Temperature":[32,45,30]})
```

```python
df5
```
