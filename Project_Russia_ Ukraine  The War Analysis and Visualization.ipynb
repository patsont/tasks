###  Use line charts/plot plot 1 by 1 and all in one to provide a visual representation of the loss of air vehicles, such as aircrafts and helicopters with all the possible parameters of sns.lineplot

```python
d1['date']=pd.to_datetime(d1['date'])
```

```python
d1.info()
```

```python
d1_monthly=d1.resample('M',on='date').sum(numeric_only=True)
```

```python
plt.figure(figsize=(8,3))
sns.lineplot(x=d1_monthly.index,y=d1_monthly.aircraft,label='Aircraft')
plt.title('No of Aircraft by Month')
plt.xlabel('Month')
plt.ylabel('Aircraft')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,3))
sns.lineplot(x=d1_monthly.index,y=d1_monthly.helicopter,label='Helicopter')
plt.title('No of Helicopter by Month')
plt.xlabel('Month')
plt.ylabel('Helicopter')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(18,10))
plt.subplot(3,2,1)
sns.lineplot(x=d1_monthly.index,y=d1_monthly.aircraft,label='Aircraft')
plt.title('No of Aircraft by Month')
plt.xlabel('Month')
plt.ylabel('Aircraft')
plt.subplot(3,2,2)
sns.lineplot(x=d1_monthly.index,y=d1_monthly.helicopter,label='Helicopter')
plt.title('No of Helicopter by Month')
plt.xlabel('Month')
plt.ylabel('Helicopter')
plt.legend()
plt.tight_layout()
plt.show()
```

### Use Sns.heatmap along with merge and corr method. Find out how each column is connected to the others since correlations are nothing more than determining how closely two variables are related . Since a correlation is nothing more than determining how closely two variables are related, find out how each column is related to the others by using corr method.

```python
d1=pd.read_csv("Dataset 1 (1).csv")
d2=pd.read_csv("Dataset 2 (1).csv")
```

```python
d_merge= pd.merge(d1,d2,on='day')
```

```python
d_merge.head()
```

```python
d_merge.info()
```

```python
d_num=d_merge.select_dtypes(include=['int64','float64'])
```

```python
d_num.head()
```

```python
d_num.corr()
```

### You can plot heatmap to identify the correlation by using heatmap method

```python
plt.figure(figsize=(22,12))
sns.heatmap(d_num.corr(),cmap='coolwarm',vmin=-1,vmax=1,annot=True,
            fmt='0.2f',annot_kws={"fontsize":14,"fontweight":"bold","fontfamily":"serif"},linewidth=1,linecolor="black")
plt.xticks(rotation=45)
plt.title('Correlation Heatmap')
plt.show()
```

### Find the loss of tanks and APC through the Bar plot.

```python
d1=pd.read_csv("Dataset 1 (1).csv")
```

```python
d1.info()
```

```python
d1['date']=pd.to_datetime(d1['date'])
```

```python
d1_month=d1.resample('M',on='date').sum(numeric_only=True)
```

```python
plt.figure(figsize=(10,3))
sns.barplot(d1_month, x=d1_month.index.strftime("%Y-%m"), y=d1_month.tank, color='yellow',label='Tank')
plt.xticks(rotation=45)
plt.title('Loss of Tanks by Month')
plt.ylabel('Frequency')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.barplot(d1_month,x=d1_month.index.strftime("%Y-%m"), y=d1_month.APC, color='green',label='APC')
plt.xticks(rotation=45)
plt.title('Loss of APC by Month')
plt.ylabel('Frequency')
plt.legend(loc='upper left')
plt.show()
```
