```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://www.sunflex.com.sg/'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
headings = soup.find_all('h2') #find_all(), find(), select(), etc.
for heading in headings:
    print(heading.text) # Print the text of each <h2> tag
```

```python
# Find all <div> tags
divs = soup.find_all('div')

# find <a> tags
for div in divs:
    a_tags = div.find_all('a')
    for a in a_tags:
        print(a.text)  # Print the text of each <a> tag
        print(a['href'])  # Print the href attribute of each <a> tag
```

```python
# Find all <a> tags within <div> tags with class="row", like this <div class="row">
a_tags_in_div = soup.select('div.row a')
for a in a_tags_in_div:
    print(a.text, a['href'])
```

```python
# Find all <footer> tags within with class="footer", like this <footer class="footer">
a_tags_in_div = soup.select('footer.footer')
for a in a_tags_in_div:
    print(a.text)
```
